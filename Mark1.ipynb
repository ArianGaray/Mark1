```python
import os
import pandas as pd
import glob

# Carpeta donde están los archivos
carpeta = r'C:\Users\User\Desktop\Automatización'

# Buscar todos los archivos .txt en la carpeta
archivos_txt = glob.glob(os.path.join(carpeta, '*.txt'))

# Lista para guardar los DataFrames
dataframes = []

# Procesar cada archivo
for archivo in archivos_txt:
    # Leer el archivo delimitado por "|"
    df = pd.read_csv(archivo, delimiter='|', encoding='latin1')
    
    # Convertir la columna 'PAGO_LINK' (DNI) a texto y completar con ceros a la izquierda si es necesario
    df['PAGO_LINK'] = df['PAGO_LINK'].astype(str).str.zfill(8)  # Asegurar que el DNI tenga 8 dígitos
    
    # Agregar el DataFrame a la lista
    dataframes.append(df)

# Concatenar todos los DataFrames, ignorando las cabeceras duplicadas
df_concatenado = pd.concat(dataframes, ignore_index=True)

# Seleccionar las columnas que necesitamos: DNI (PAGO_LINK) y los teléfonos
df_telefonos = df_concatenado[['PAGO_LINK', 'FCTEL1', 'FCTEL2', 'FCTEL3', 'FCTEL4', 'FCTEL5']]

# Unpivot los teléfonos para que estén en una sola columna
df_telefonos_melted = df_telefonos.melt(id_vars='PAGO_LINK', value_vars=['FCTEL1', 'FCTEL2', 'FCTEL3', 'FCTEL4', 'FCTEL5'],
                                        var_name='Tipo_Telefono', value_name='Teléfono')

# Eliminar filas donde el teléfono sea NaN
df_telefonos_melted = df_telefonos_melted.dropna(subset=['Teléfono'])

# Convertir los teléfonos a enteros para eliminar el .0, y luego a texto para que se guarden bien en Excel
df_telefonos_melted['Teléfono'] = df_telefonos_melted['Teléfono'].astype(int).astype(str)

# Filtrar y eliminar los teléfonos que no comienzan con "9" o que sean "999999999"
df_telefonos_melted = df_telefonos_melted[df_telefonos_melted['Teléfono'].str.startswith('9')]
df_telefonos_melted = df_telefonos_melted[df_telefonos_melted['Teléfono'] != '999999999']

# Eliminar la columna 'Tipo_Telefono' que no es necesaria
df_final = df_telefonos_melted[['PAGO_LINK', 'Teléfono']]

# Crear la columna "Concatenado" con el formato DNI-TELEFONO
df_final['Concatenado'] = df_final['PAGO_LINK'] + '-' + df_final['Teléfono']

# Ordenar los DNI de menor a mayor
df_final = df_final.sort_values(by='PAGO_LINK')

# Exportar el DataFrame agrupado a un archivo Excel en la carpeta indicada
ruta_salida = os.path.join(carpeta, 'dni_telefonos_concatenados.xlsx')
df_final.to_excel(ruta_salida, index=False)  # Exportar sin el índice

print(f"Archivo exportado exitosamente en: {ruta_salida}")
```
